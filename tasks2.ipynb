```python
# Step 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load Dataset (using the uploaded CSV file)
df = pd.read_csv('gender_submission.csv')

# Step 3: Basic Info (trying to understand what is in the data)
print("First 5 rows:")
print(df.head())

print("\nShape of data:", df.shape)
print("\nData types:")
print(df.dtypes)

# Step 4: Check missing values
print("\nMissing values:")
print(df.isnull().sum())

# Step 5: Count plot of Survived (how many survived vs not)
sns.countplot(data=df, x='Survived', palette='pastel')
plt.title('Passenger Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Step 6: Pie chart for survival
survived_counts = df['Survived'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(survived_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Survival Rate')
plt.show()

# Step 7: Print some observations (like an intern would!)
print("\n Observations:")
print("➡ Total passengers:", df.shape[0])
print("➡ Survivors:", df['Survived'].sum())
print("➡ Survival rate: {:.2f}%".format(df['Survived'].mean() * 100))
```
