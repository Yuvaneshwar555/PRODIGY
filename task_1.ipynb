```python
# Step 1: Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Creating a simple dataset manually
# (This is like what I learned from YouTube and tried to make on my own )
data = {
    'Age_Group': ['0-14', '15-24', '25-54', '55-64', '65+'],
    'Population_Millions': [30, 20, 50, 15, 10]
}

# Step 3: Converting it into a DataFrame
df = pd.DataFrame(data)

# Step 4: Plotting the Bar Chart
plt.figure(figsize=(8,5))
plt.bar(df['Age_Group'], df['Population_Millions'], color='skyblue', edgecolor='black')
plt.title('Population Distribution by Age Group', fontsize=14)
plt.xlabel('Age Groups', fontsize=12)
plt.ylabel('Population (in Millions)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Step 5: Showing the Plot
plt.show()
```
